```python
import random
```

```python
lives = 5
```

```python
def guess(word, placeholder, lives=5, total=5):
    g = input(f"({lives}/{total}) Guess a letter: ")
    if g in word:
        print("Good guess!")
        placeholder = [x for x in placeholder]
        for i in range(len(word)):
            if word[i] == g:
                placeholder[i] = g
        return "".join(placeholder), lives
    else:
        print("Bad guess!")
        return placeholder, lives-1
```

```python
def hangman():
    print("HANGMAN")
    lives,total = 5,5
    word = random.choice(WORDLIST)
    m = len(word)
    placeholder = "-"*m
    while lives>0:
        print("\n",placeholder)
        placeholder, lives = guess(word, placeholder, lives)
        if placeholder==word:
            print("\nYou Won!")
    print("\nYou Lost")
    
```
